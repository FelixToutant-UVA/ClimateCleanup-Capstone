```python
import json
import pandas as pd

# read file
with open("bomen_test.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
display(df)

# dutch provinces
provinces = [
    'Drenthe', 'Flevoland', 'Friesland', 'Gelderland', 'Groningen',
    'Limburg', 'Noord-Brabant', 'Noord-Holland', 'Overijssel',
    'Utrecht', 'Zeeland', 'Zuid-Holland'
]
df_nl = df[df["provincie"].isin(provinces)]

# age to integer
df_nl["leeftijd plot"] = pd.to_numeric(df_nl["leeftijd plot"], errors="coerce")

# Filter alleen positieve leeftijden
df_nl = df_nl[df_nl["leeftijd plot"] > 0]

# average age
average_age = df_nl["leeftijd plot"].mean()

print(f"Average age: {average_age:.1f} jaar")
```
